```python
res1 = req.get('https://www.melon.com')
res1
# 서버는 브라우저를 통해서만 접근해야된다
# header에 User-Agent를 지정하고 접근하면 해결 가능
# header는 딕셔너리 형식으로 사용
```

```python
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36 Edg/105.0.1343.33'}
```

```python
res1 = req.get('https://www.melon.com/', headers=header)
res1
```

```python
res1.text
```
