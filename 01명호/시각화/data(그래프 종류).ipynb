```python
# seaborn : 시각화를 쉽게 만들어주는 lib
# jump, minitab 그래프랑 똑같음
# 머신러닝에서 많이 사용
import seaborn as sns
```

```python
# lib자체에 test data들이 많이 있음
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker',palette='Set3')
# hue : 추가로 data 분류하고 싶을 때 사용
plt.ylim(0,80)
plt.show()
```

```python
penguins=sns.load_dataset('penguins')
penguins.head()
```

```python
sns.histplot(data=penguins,x='flipper_length_mm',hue='species',multiple='stack')
# multiple='stack' : 그래프 안겹치게 만들어 줌
plt.show()
```

```python
# displot : distribution을 여러 subplot으로 나누어 출력
sns.displot(data=penguins,x='flipper_length_mm',hue='species',col='species')
# col을 기준으로 그래프를 나눠 줌
plt.show()
```

```python
# countplot: 범주형 속성을 가지는 data의 히스토그램
sns.countplot(data=penguins,y='species',hue='sex')
plt.show()
```

```python
# boxplot : outliner 찾을 때 편함
sns.boxplot(x='species',y='body_mass_g',hue = 'species', data=penguins,palette="Set3")
plt.show()
```

```python
# violinplot : 전체 data 분포 확인
sns.violinplot(x='species',y='body_mass_g',hue = 'species', data=penguins, palette="Set3")
plt.show()
```

```python
# lineplot : 3개 분포 비교해서 보고 싶을 때 사용해도 될 듯
sns.lineplot(x='bill_depth_mm',y='bill_length_mm',hue = 'species', data=penguins)
plt.show()
```

```python
# scatterplot : 점 분포
sns.scatterplot(x='bill_depth_mm',y='bill_length_mm',hue = 'sex', data=penguins)
plt.show()
```

```python
# pairplot : dataa 각 feature들 사이의 모든 관계를 표시
# 라인별 * 불량별 data 뽑을 때 사용
sns.pairplot(data=penguins,hue='island')
plt.show()
```
