```python
midtest_avg = pd.Series([75,80,90,95,70], index = ['국어','영어','수학','과학','사회'])
midtest_avg.name = '중간고사 성적'
midtest_avg.index.name = '과목'
midtest_avg
```

```python
midtest_avg[:3]
```

```python
midtest_avg.drop(['과학','사회'])
```
